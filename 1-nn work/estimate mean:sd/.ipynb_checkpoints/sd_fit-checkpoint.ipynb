```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.optimize import curve_fit
from scipy.optimize import differential_evolution
```

```python
df = pd.read_csv('sd_data.csv')
```

```python
df[df['p'] == 2]
```

```python
def fn(x, m, k, C):
    return m * x ** k + C
```

```python
# fit (m, k, C) for all p

p_uniq = list(set(df['p']))
fitted_coef = np.empty((len(p_uniq),3))

for i in range(len(p_uniq)):
    p = p_uniq[i]
    
    df_temp = df[df['p'] == p]
    
    x = np.array(df_temp.iloc[:,1], dtype=float)
    y = np.array(df_temp.iloc[:,3], dtype=float)

    # curve fit the test data
    fittedParameters, pcov = curve_fit(fn, x, y, bounds=((0, 0, -np.inf), (np.inf, np.inf, np.inf)))

    fitted_coef[i,] = fittedParameters

    modelPredictions = fn(x, *fittedParameters) 

    absError = modelPredictions - y

    Rsquared = 1.0 - (np.var(absError) / np.var(y))
    print('R-squared:', Rsquared)
    
coef_df = pd.DataFrame({"p" : p_uniq, 
                    "m": fitted_coef[:,0],
                    "k": fitted_coef[:,1],
                    "C": fitted_coef[:,2],
                   }).sort_values(by=['p'])
```

```python
coef_df
```

```python
def plot_pred(p):
    data = df[df['p'] == p]
    
    xModel = np.linspace(2, 500)
    yModel = fn(xModel, *coef_df[coef_df['p'] == p].iloc[0,1:4])
    
    plt.scatter(data.iloc[:,1], data.iloc[:,3])
    plt.plot(xModel, yModel)
    
    plt.show()
    plt.close()
```

```python
plot_pred(2)
```

```python
k_avg = np.average(coef_df['k'])
```
